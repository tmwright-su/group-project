```python
import pandas as pd
import requests as req 
import json 
import sys

import warnings
warnings.filterwarnings('ignore')

books_db = pd.read_csv('books.csv') # converting out data into a dataframe 

#this function output a random sample from our database
# The word lucky works similarly to google's "I'm feeling Lucky". Lucky gives you random book sample similiary to how google I'm feeling lucky gives you random info"


def lucky_output(start_search):
    
    if start_search == 'Lucky':
        sample = books_db.sample(10)
    return(sample)
  
genre_list = books_db['Genre'].tolist() #converting the Genre column of our program into a list 

# this function prints out unique values in our Genre column so no Genre repeats
def unique(genre_list):
    list_set = set(genre_list)
    unique_list = list(list_set)
    for x in unique_list:
        print(x, end=',')

# this function ensures that when the user types in a genre, it's a genre from the genre column
def genre(genre_pref):
    genre = books_db['Genre']== genre_pref
    genre_db = books_db[genre]
    genre_db

start_search = str(input("\n Enter Lucky or Custom:")).title()

if start_search == "Lucky":
    lucky_output(start_search)
    sys.exit(lucky_output(start_search)) # if the user enters Lucky the program ends and gives a sample from data
      
elif start_search == "Custom":
    print(unique(genre_list)) # prints out list of genre's so the user knows what to choose from
else:
    sys.exit("Invalid Entry enter Lucky or Custom.")
    
    
genre_pref = str(input("\n Enter Your Favorite Genre From The List of Genres:")).title()
genre = books_db['Genre']== genre_pref   # genre entered is equalivant to the genre in books database
genre_db = books_db[genre]

#if genre_pref != genre_db[genre_db["Genre"]]: 
    #sys.exit('Not in Genre list')
    
#RATING is calling from genre database 

try:
    rate_pref = int(input("On a scale of 1-4, how important is the rate of the book? (1 being low importance, 4 being high importance): "))   
except ValueError:
    print("Enter integers only")
    
try:
    if rate_pref==1: 
        rate_db = genre_db

    elif rate_pref==2:
        rate_db = genre_db[ genre_db['Rating']>=2.0]

    elif rate_pref==3:
        rate_db = genre_db[ genre_db['Rating']>=3.0]

    elif rate_pref==4:
        rate_db = genre_db[ genre_db['Rating']>=4.3]

    else: 
        print('Invalid Entry \n')

except NameError:
    print('Program will not work at intended until a numeric value!')


# PAGES is calling from our rating database 

try:
    pages_pref = int(input("On a 1-4 scale, rate how long you would like the book to be: \n"))
except ValueError:
    print("Enter integers only")

try:
    if pages_pref==1:
        pages_db = rate_db

    elif pages_pref==2:
        pages_db = rate_db[ rate_db['PAges']> 200]

    elif pages_pref==3:
        pages_db = rate_db.query('200 <= PAges <= 400')

    elif pages_pref==4:
        pages_db = rate_db[ rate_db['PAges']> 400]

    else: 
        print('Invalid Entry')
except NameError:
    print("Program will not work as intended until you enter a numeric value!")

    
#YEAR is calling from our pages database. In the end we created one databasee based of the responses of the users answers for the previous database. 

try:
    year_pref = int(input("On a scale of 1-4, rate how old or new you would like the book to be: "))
except ValueError:
    print("Enter integers only \n")
    
try:
    if year_pref==1:
        year_db = pages_db

    elif year_pref==2:
        year_db = pages_db.query('1800 <= Year <= 1920')

    elif year_pref==3:
        year_db = pages_db.query('1921 <= Year <= 1990')

    elif year_pref==4:
        year_db = pages_db[ pages_db['Year']>1990] 
except NameError:
    print("Program will not work as intended until you enter a numeric value!")
    
print("\n Your customized Booklist is below ")
year_db.head(10)
```
